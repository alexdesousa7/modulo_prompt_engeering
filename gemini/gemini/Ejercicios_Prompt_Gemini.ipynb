## Ejercicio 5: Cadena de Pensamiento (Chain-of-Thought)

**Objetivo:** Animar al modelo a "pensar en voz alta" o desglosar un problema en pasos para llegar a una solución más precisa, especialmente en problemas de lógica o matemáticos.

**Instrucciones:** Formula un `prompt` para un problema de matemáticas que requiera varios pasos para resolver. Por ejemplo, "Dos trenes salen de dos ciudades diferentes a la misma hora. El tren A viaja a 80 km/h y el tren B a 100 km/h. ¿A qué distancia estarán el uno del otro después de 2 horas? Desglosa tu razonamiento paso a paso antes de dar la respuesta final."

```python
problema = (
    "Supón que la Tierra, con una masa de aproximadamente 5.972 × 10^24 kg, "
    "se convierte en un agujero negro. Utiliza la fórmula del radio de Schwarzschild: "
    "r_s = (2GM) / c^2, donde G es la constante de gravitación universal (6.674 × 10^-11 m^3 kg^-1 s^-2) "
    "y c es la velocidad de la luz en el vacío (3 × 10^8 m/s). "
    "Calcula paso a paso el radio de Schwarzschild para la Tierra, "
    "explica cada paso de tu razonamiento y compara el resultado con el tamaño de un objeto cotidiano."
)

prompt = (
    f"Resuelve el siguiente problema utilizando una cadena de razonamiento paso a paso:\n\n"
    f"{problema}\n\n"
    "Primero, explica qué es el radio de Schwarzschild y para qué sirve. "
    "Luego, sustituye los valores numéricos en la fórmula y realiza los cálculos intermedios. "
    "Finalmente, da el resultado final en metros y compáralo con el tamaño de un objeto cotidiano para dar contexto."
)

model="gemini-3-flash-preview" # Si se acaba la cuota ir probando con otros modelos
response = client.models.generate_content(model=model, contents=prompt)

display(Markdown(response.text))
```

## Ejercicio 6: Prompt de Rol Múltiple

**Objetivo:** Pedirle al modelo que responda a una situación desde la perspectiva de varios roles diferentes.

**Instrucciones:** Formula un prompt para que el modelo explique cómo afrontar un conflicto en el trabajo, primero como jefe, luego como compañero y finalmente como responsable de recursos humanos. El modelo debe dar una respuesta diferenciada para cada rol. Plantea tú mismo el escenario del conflicto.


```python
# Escenario del conflicto planteado
escenario = (
    "En una empresa tecnológica, dos empleados han tenido un desacuerdo importante durante una reunión. "
    "Uno de ellos interrumpió repetidamente al otro cuando explicaba cómo resolver un problema técnico del proyecto de IAGen en el que trabajan, lo que generó tensión y malestar en el equipo."
    "El conflicto ha afectado la colaboración y el ambiente laboral."
)

# Prompt para el modelo, solicitando respuestas diferenciadas por rol
prompt = (
    f"Escenario: {escenario}\n\n"
    "Explica cómo afrontar este conflicto en el trabajo desde tres perspectivas diferentes:\n"
    "1. Como jefe: ¿Qué acciones tomarías para resolver el conflicto y restaurar el ambiente de trabajo?\n"
    "2. Como compañero: ¿Cómo apoyarías a tus colegas y contribuirías a una solución positiva?\n"
    "3. Como responsable de recursos humanos: ¿Qué medidas implementarías para gestionar el conflicto y prevenir situaciones similares en el futuro?\n\n"
    "Da una respuesta diferenciada y estructurada para cada rol."
)
model="gemini-2.5-flash-lite" # Si se acaba la cuota ir probando con otros modelos
response = client.models.generate_content(model=model, contents=prompt)
display(Markdown(response.text))
```

## Ejercicio 8: Prompt de Evaluación Crítica

**Objetivo:** Solicitar al modelo que analice críticamente un argumento o texto, identificando fortalezas, debilidades y posibles sesgos.

**Instrucciones:** Escribe un prompt para que el modelo evalúe críticamente el ensayo generado en el ejercicio anterior, señalando al menos dos puntos fuertes, dos débiles y cualquier sesgo detectado en el argumento.

```python
texto = response.text
# Prompt para evaluación crítica del ensayo generado
prompt_evaluacion = (
    "Lee el siguiente ensayo y realiza una evaluación crítica. "
    "Identifica al menos dos puntos fuertes, dos puntos débiles y cualquier sesgo presente en el argumento. "
    "Presenta la respuesta en formato de lista Markdown, separando claramente fortalezas, debilidades y sesgos.\n\n"
    f"Ensayo:\n\"\"\"\n{texto}\n\"\"\""
)
response_evaluacion = client.models.generate_content(model=model, contents=prompt_evaluacion)
display(Markdown(response_evaluacion.text))
```
