# Ejemplo de Uso de la API de Gemini con Python

Este notebook es una guía de inicio rápido para interactuar con la API de Google Gemini usando Python.

## Capa Gratuita

Gemini cuenta con una capa gratuita muy amplia que te permite experimentar y construir aplicaciones. Los límites dependen del modelo gratuito que uses, pero aproximadamente son:

- **5/10 peticiones por minuto**
- **250k tokens por minuto**
- **20 peticiones por día**

Estos límites se han visto reducidos significativamente con la salida de Gemini-3-Flash, ya que ha aumentado mucho la cantidad de gente usando la IA de Google y no tienen suficiente capacidad de cómputo, presúntamente. Hasta Diciembre de 2025 tenías hasta 1000 peticiones al día y entre 15 y 20 peticiones por minuto.

Cuando el modelo ha llegado a su límite de peticiones o está sobrecargado en el momento de uso, devuelve un error como este `ServerError: 503 UNAVAILABLE. {'error': {'code': 503, 'message': 'The model is overloaded. Please try again later.', 'status': 'UNAVAILABLE'}}`).

## Obtén tu API Key

Para ejecutar el código en este notebook, necesitas generar una API Keys de Google.

1.  Visita [Google AI Studio](https://aistudio.google.com/app/apikey).
2.  Inicia sesión con tu cuenta de Google y crea un nuevo proyecto si es necesario (por defecto, viene creado uno llamado 'Default Gemini Project')
3.  Haz clic en "**Create API key**" para generar tu clave (asociada a un proyecto)
4.  Copia la clave generada y pégala en la variable `GOOGLE_API_KEY` en la celda de código que se encuentra a continuación.



Una vez que hayas configurado tu clave, podrás ejecutar el resto de las celdas para comunicarte con el modelo Gemini.

```python
# pip install -U -q "google-genai" instala la librería necesaria para interactuar con Google GenAI
from google import genai
from google.genai import types

from IPython.display import display, Markdown

# Configurar la API Key
GOOGLE_API_KEY='Tu API Key aquí'  # Reemplaza con tu API Key de Google GenAI
client = genai.Client(api_key=GOOGLE_API_KEY)
```

```python
response = client.models.generate_content(
    model='gemini-2.5-flash-lite',
    contents='¿Explícame el sentido de la vida?',
)
display(Markdown(response.text))
```
