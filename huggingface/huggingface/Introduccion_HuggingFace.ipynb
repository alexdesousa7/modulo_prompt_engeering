# **Introducción a HuggingFace** 🤗

### **¿Qué es HuggingFace y por qué es diferente a Groq (o a AwanLLM, API de OpenAI, API de Google AI Studio...)?**

**Groq** es un proveedor de inferencia (como OpenAI, Google AI Studio o AwanLLM). Te conectas a sus servidores y usas sus modelos alojados en su infraestructura pagando por uso, o tirando del tier gratuito haciendo llamadas via API.

**HuggingFace** es un ecosistema completo:

- Hub/Repositorio: Es el "GitHub de IA" - miles de modelos open-source
- Librerías (transformers, datasets, diffusers): Para cargar y ejecutar modelos LOCALMENTE o en tu infraestructura
- También tienes inferencia vía API como los proveedores Groq, AwanLLM, OpenAI, Google AI Studio...
- Spaces: Hosting de demos/aplicaciones
- Datasets: Miles de datasets públicos

### **Groq vs HuggingFace: Diferencias Clave**

| Aspecto | Groq (y OpenAI/Gemini) | HuggingFace |
|---------|-----------------|-------------|
| **Tipo de servicio** | Proveedor de inferencia | Ecosistema completo ML |
| **Dónde se ejecuta** | En sus servidores | En TU máquina/servidor |
| **Acceso al modelo** | Solo API (caja negra) | Código completo y pesos |
| **Costos** | Por token/uso | Gratis (solo hardware) |
| **Personalización** | Limitada a parámetros | Fine-tuning completo |
| **Dependencia** | Internet + API Key | Solo recursos locales |
| **Complejidad** | Muy simple (API) | Mayor (gestión de modelos) |

### **¿Por qué HuggingFace es tan demandado en el mercado laboral?**

1. Es el **Estándar de facto** para modelos open-source
2. **Flexibilidad total**: Descargas el modelo y lo ejecutas donde quieras

## 1. **La forma más sencilla: `pipeline()`**

`pipeline()` es una función que:
1. Descarga el modelo automáticamente

### **1.1. Generación de texto**

````python
from IPython.display import display
from IPython.core.display import Markdown
````

````python
# Pipeline simple para generar texto
# Usamos distilgpt2, que es un modelo pequeño perfecto para CPU
generador = pipeline(
    "text-generation",
    model="distilbert/distilgpt2",
    device=-1  # -1 = CPU, 0 = GPU
)

# Generamos el texto
prompt = "The future of artificial intelligence is"
respuesta = generador(
    prompt,
    max_length=100,
    num_return_sequences=1,
    temperature=0.7
)

content = respuesta[0]['generated_text']
display(Markdown(content))
````

### **1.2. Análisis de Sentimientos**

````python
# Pipeline para análisis de sentimientos
clasificador = pipeline(
    "sentiment-analysis",
    model="distilbert/distilbert-base-uncased-finetuned-sst-2-english",
    device=-1
)

textos = [
    "I love this product! It's amazing!",
    "This is terrible, worst purchase ever.",
    "It's okay, nothing special."
]

for texto in textos:
    respuesta = clasificador(texto)
    display(Markdown(f"**Texto:** {texto}"))
    display(Markdown(f"**Sentimiento:** {respuesta[0]['label']} (confianza: {respuesta[0]['score']:.2%})\n"))
````

### **1.3. Question Answering (Responder preguntas sobre un contexto)**

````python
# Pipeline de Question Answering
qa_model = pipeline(
    "question-answering",
    model="distilbert/distilbert-base-cased-distilled-squad",
    device=-1
)

contexto = """
HuggingFace es una solución proveedora de inferencia de LLMs y un entorno
donde acceder a modelos opensource para usarlos de forma local.
Fue fundada en 2016 en la ciudad de París, Francia.
La compañía se ha convertido en el hub principal para modelos de machine learning open-source.
En 2021, recaudaron 40 millones de dólares en financiación Serie B.
"""

preguntas = [
    "¿Cuándo fue fundada HuggingFace?",
    "¿Dónde fue fundada?",
    "¿Cuánto dinero recaudaron en 2021?"
]

for pregunta in preguntas:
    resultado = qa_model(question=pregunta, context=contexto)
    display(Markdown(f"**P:** {pregunta}"))
    display(Markdown(f"**R:** {resultado['answer']} (confianza: {resultado['score']:.2%})\n"))
````

## **2. Control sobre el modelo: Tokenizer + (Embedding) + Model**

Para tener control total, necesitas entender dos componentes:

### **Tokenizer**
Convierte texto -> números que el modelo entiende

### **Model**
Hace las predicciones/generaciones

Es más complejo pero te da **control total** sobre cada parámetro.

### **2.1. Entendiendo el Tokenizer**

````python
tokenizer = AutoTokenizer.from_pretrained("distilbert/distilgpt2")

texto = "Hello, how are you? My name is Alonso and I am learning HuggingFace. I love natural language processing! Also, I enjoy hiking and exploring new places."

# Veamos cómo tokeniza
tokens = tokenizer.tokenize(texto)
print(f"Texto original: {texto}")
print(f"Tokens: {tokens}")

# Veamos que ID asigna a cada elemento (encoding)
ids = tokenizer.encode(texto)
print(f"IDs: {ids}")

# Veamos como reconstruir el texto a partir de los IDs (decoding)
texto_reconstruido = tokenizer.decode(ids)
print(f"Reconstruido: {texto_reconstruido}")
````

Ese carácter raro que véis Ġ representa un espacio en blanco en la tokenización. Es la forma en que el tokenizer marca: "esta palabra viene después de un espacio". Es parte del sistema Byte-Pair Encoding (BPE) que usa GPT-2.

- 'Hello' = inicio de frase, sin espacio delante
- 'Ġhow' = la palabra "how" con un espacio antes
- 'Ġprocessing' = la palabra "processing" con espacio antes

Esto permite al modelo diferenciar entre "unhappy" y "un happy" (con espacio).

### **2.2. Embeddings: La capa oculta después de los IDs**

Después de convertir el texto en IDs, el modelo convierte cada ID en un **vector denso** (embedding).

**Flujo completo:**
1. Texto: `"Hello world"`
2. Tokens: `['Hello', 'Ġworld']`
3. IDs: `[15496, 995]`
4. **Embeddings**: Cada ID → vector de N dimensiones (ej: 768 en GPT-2)

Los embeddings capturan el "significado" del token en un espacio vectorial multidimensional.

````python
# Para acceder a la matriz de embeddings del modelo
model = AutoModelForCausalLM.from_pretrained("distilbert/distilgpt2")
embedding_layer = model.transformer.wte  # wte = Word Token Embeddings (en GPT-2)

print("-" * 70)
print("Matriz de embeddings del modelo")
print("-" * 70)
print(f"Shape de la matriz: {embedding_layer.weight.shape}")
print(f"  └─ {embedding_layer.weight.shape[0]:,} tokens en el vocabulario")
print(f"  └─ {embedding_layer.weight.shape[1]} dimensiones por embedding")

# Veamos el embedding de un token específico
token_id = 15496  # "Hello"
embedding_vector = embedding_layer.weight[token_id]

print(f"\nEmbedding del token 'Hello' (ID: {token_id})")
print(f"Dimensiones: {embedding_vector.shape}")
print(f"Primeros 10 valores: {embedding_vector[:10].tolist()}")
print(f"Tipo de dato: {embedding_vector.dtype}")
````

````python
# Veamos embeddings de varios tokens
texto = "Hello world AI"
tokens = tokenizer.tokenize(texto)
ids = tokenizer.encode(texto, add_special_tokens=False)

print("-" * 70)
print("Embeddings de una frase completa")
print("-" * 70)
print(f"Texto: '{texto}'\n")

for token, token_id in zip(tokens, ids):
    embedding = embedding_layer.weight[token_id]
    print(f"Token: '{token}' (ID: {token_id})")
    print(f"  Embedding shape: {embedding.shape}")
    print(f"  Primeros 5 valores: {embedding[:5].tolist()}")
    print(f"  Norma L2: {embedding.norm().item():.4f}")
    print()
````

### **2.3. Similaridad entre embeddings**

Los embeddings permiten calcular similaridad semántica entre palabras usando distancia coseno:

````python
import torch.nn.functional as F

# Función para calcular similitud del coseno
def cosine_similarity(emb1, emb2):
    return F.cosine_similarity(emb1.unsqueeze(0), emb2.unsqueeze(0)).item()

# Comparemos similitud entre palabras
palabras = ["king", "queen", "man", "woman", "apple"]

print("-" * 70)
print("Similitud del coseno entre embeddings")
print("-" * 70)

# Obtenemos embeddings de las palabras
embeddings_dict = {}
for palabra in palabras:
    token_ids = tokenizer.encode(palabra, add_special_tokens=False)
    if len(token_ids) > 0:
        token_id = token_ids[0]
        embeddings_dict[palabra] = embedding_layer.weight[token_id]

# Calculamos su similitud
print("\nSimilaridad entre palabras:\n")
for i, palabra1 in enumerate(palabras):
    for palabra2 in palabras[i+1:]:
        if palabra1 in embeddings_dict and palabra2 in embeddings_dict:
            sim = cosine_similarity(embeddings_dict[palabra1], embeddings_dict[palabra2])
            print(f"'{palabra1}' ↔ '{palabra2}': {sim:.4f}")
````

### **¿Se pueden modificar los embeddings?**

**SÍ, puedes modificarlos** (útil para fine-tuning o inyectar conocimiento):

```python
# Modificar el embedding de un token
token_id = 15496
embedding_layer.weight[token_id] = torch.randn(768)  # Nuevo embedding aleatorio

# O actualizar durante el entrenamiento
# Los embeddings son parámetros entrenables del modelo
```

**Casos de uso:**
1. **Fine-tuning**: Los embeddings se ajustan automáticamente durante el entrenamiento
2. **Vocabulary expansion**: Añadir nuevos tokens (ej: nombres propios, jerga técnica)
3. **Embedding initialization**: Inicializar embeddings de tokens nuevos
4. **Transfer learning**: Copiar embeddings de un modelo a otro

**Pero**: Modificar embeddings arbitrariamente puede romper el modelo. Solo hazlo si sabes lo que haces o estás fine-tuning.

### **2.4. Cargando y usando el Modelo**

````python
model = AutoModelForCausalLM.from_pretrained("distilbert/distilgpt2")

# Pad_token (importante para evitar warnings)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

prompt = "The meaning of life is"
inputs = tokenizer(prompt, return_tensors="pt", padding=True)

print("Input preparado:")
print(f"  input_ids shape: {inputs['input_ids'].shape}")
print(f"  input_ids: {inputs['input_ids']}")
````

### **2.5. Generación de texto con control total**

````python
# Similar a lo que haciamos con el servicio groq vía API, interactuamos con los parámetros del modelo
outputs = model.generate(
    inputs["input_ids"],
    attention_mask=inputs["attention_mask"],
    max_length=50,              # Similar a max_completion_tokens
    num_return_sequences=1,     # Cuántas respuestas generar
    temperature=0.7,
    top_p=0.9,
    do_sample=True,             # True = usa temperature, False = greedy
    pad_token_id=tokenizer.eos_token_id,
    no_repeat_ngram_size=2      # Evita repeticiones
)

# Decodificamos
texto_generado = tokenizer.decode(outputs[0], skip_special_tokens=True)
display(Markdown(texto_generado))
````

### **4.2. ¿Dónde se descargan los modelos?**

Cuando usas `AutoModelForCausalLM.from_pretrained()`, el modelo **se descarga en local**.

**Ubicación por defecto:**
- **Windows**: `C:\Users\<usuario>\.cache\huggingface\hub\`
- **Linux/Mac**: `~/.cache/huggingface/hub/`

**¿Qué significa esto?**
- La **primera vez** descarga el modelo (tarda según el tamaño)
- Las **siguientes veces** lo carga desde el disco (mucho más rápido)
- **Gastas espacio en disco** (no RAM hasta que lo cargas en memoria)
- Puedes **trabajar offline** una vez descargado

**Ejemplo de tamaños:**
- distilgpt2: ~330 MB
- TinyLlama-1.1B: ~2.2 GB
- Phi-2: ~5.2 GB
- Llama-2-7B: ~13 GB

````python
# Para ver dónde se guardan los modelos
import os
from pathlib import Path

cache_dir = Path.home() / ".cache" / "huggingface" / "hub"
print(f" Directorio de caché de HuggingFace:")
print(f"   {cache_dir}")
print(f"\n¿Existe? {cache_dir.exists()}")

if cache_dir.exists():
    modelos = list(cache_dir.glob("models--*"))
    print(f"\n Modelos descargados: {len(modelos)}")
    for modelo in modelos[:5]:  # Primeros 5
        nombre = modelo.name.replace("models--", "").replace("--", "/")
        # Calcular tamaño
        size_mb = sum(f.stat().st_size for f in modelo.rglob('*') if f.is_file()) / (1024**2)
        print(f"   - {nombre}: {size_mb:.1f} MB")
````

### **4.3. Pipeline vs AutoModel: ¿Cuál usar?**

| Aspecto | `pipeline()` | `AutoModelForCausalLM` + `AutoTokenizer` |
|---------|--------------|------------------------------------------|
| **Facilidad** | Muy fácil | Más complejo |
| **Control** | Limitado | Total |
| **Configuración** | Automática | Manual |
| **Uso** | Tareas estándar | Personalización avanzada |
| **Cuando usar** | Prototipado rápido | Fine-tuning, producción, embeddings |

**Pipeline internamente usa AutoModel**, solo que lo hace automáticamente:
```python
# Esto:
pipeline("text-generation", model="gpt2")

# Es equivalente a (simplificado):
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")
# + lógica adicional de pipeline
```

**¿Cuándo usar cada uno?**

**Usa `pipeline()`** cuando:
- Quieres resultados rápidos
- Tarea estándar (clasificación, Q&A, generación)
- Estás prototipando

**Usa `AutoModel`** cuando:
- Necesitas acceder a embeddings
- Vas a hacer fine-tuning
- Quieres control total sobre parámetros
- Necesitas modificar la arquitectura
- Implementas lógica custom

### **4.4. Caché vs save_pretrained(): ¿Cuál es la diferencia?**

#### **Caché automática** (`~/.cache/huggingface/`)
- **NO se borra sola** - es persistente
- Se crea **automáticamente** al hacer `from_pretrained()`
- La próxima vez que cargues el mismo modelo, lo toma de aquí (o sea,que es rápido)
- Solo tú puedes borrarlo manualmente

#### **`save_pretrained()` - Guardar en ubicación específica**
Esto es para casos diferentes:

**¿Cuándo usarlo?**
1. **Después de fine-tuning**: Has entrenado el modelo con tus datos
2. **Versionar modelos**: Guardar v1, v2, v3 en carpetas distintas
3. **Compartir con otros**: Dar el modelo a un colega
4. **Organizar proyectos**: Cada proyecto con sus modelos
5. **Mover a producción**: Copiar a un servidor
6. **Hacer modificaciones**: Backup antes de cambiar embeddings

````python
# CASO DE USO: Has hecho fine-tuning y quieres guardar tu versión personalizada
save_path = "./mi_modelo_personalizado"

# Guardas el modelo + tokenizer
model.save_pretrained(save_path)
tokenizer.save_pretrained(save_path)
print(f"Modelo guardado en: {save_path}")
print(f"Ahora puedes compartir esta carpeta o moverla a otro servidor/máquina")

# Cargar tu versión personalizada (no la original del Hub)
modelo_cargado = AutoModelForCausalLM.from_pretrained(save_path)
tokenizer_cargado = AutoTokenizer.from_pretrained(save_path)
print("Modelo personalizado cargado desde disco local")
````
